## Semana I - Limpeza dos dados trazidos de uma API

```python
import pandas as pd
```

### Lendo o arquivo no formato JSON

```python
dados = pd.read_json('/content/Telco-Customer-Churn.json')
```

```python
dados.head(10)
```

```python
type(dados)
```

### Normalizando os dados - Separando os dados de cada coluna

```python
dados_customer = pd.json_normalize(data = dados.customer)
dados_customer
```

```python
dados_phone = pd.json_normalize(data = dados.phone)
dados_phone
```

```python
dados_internet = pd.json_normalize(data = dados.internet)
dados_internet
```

```python
dados_account = pd.json_normalize(data = dados.account)
dados_account
```

```python
dados_ID = dados[['customerID',	'Churn']]
dados_ID
```

### Juntando todos os DataFrames em apenas um

```python
df = pd.concat([dados_ID, dados_customer, dados_phone, dados_internet, dados_account], axis = 1)
df
```

### Ampliando a visualização do número de colunas

```python
pd.set_option("display.max_columns",30)
df
```

### Traduzindo as Colunas

```python
df = df.rename(columns={'customerID': 'IDcliente', "gender": "sexo", "SeniorCitizen": "idoso", "Partner": "parceiro", "Dependents": "dependentes", "tenure": "tempo_contrato", "PhoneService": "servico_telefonico",
                   "MultipleLines": "multiplas_linhas", "InternetService": "servico_internet", "OnlineSecurity": "segurança_online", "OnlineBackup": "backup_online", "DeviceProtection": "protecao_dispositivo",
                   "TechSupport": "suporte_tecnico", "StreamingTV": "TV_a_cabo", "StreamingMovies": "streaming_filmes", "Contract": "tipo_contrato", "PaperlessBilling": "conta_online", "PaymentMethod": "forma_pagamento",
                   "Charges.Monthly": "gastos_mes", "Charges.Total": "gastos_total"})
```

### Verificando Inconsistencias nos Dados

```python
colunas = df.columns.tolist()
for coluna in colunas:
  print(f'A coluna {coluna} tem as seguintes informações: {df[coluna].unique()}')
```

Foi verificado que a coluna Churn contem informações vazias e a coluna gastos_total está no formato object. Corrigiremos essas inconsistencias a seguir. 

### Analizando a coluna Churn

```python
selecao = df['Churn'] == ''
df[selecao].shape
```

```python
df[selecao].head(10)
```

```python
# Apagando as linhas que contém o Churn = ''
selecao = df['Churn'] != ''
df = df[selecao]
```

```python
df.Churn.unique()
```

```python
df.info()
```

### Analizando a coluna gastos_total

```python
df['gastos_total']
```

#### Mudando o tipo de objeto para numerico 

```python
df['gastos_total'] = pd.to_numeric(df['gastos_total'], errors='coerce')
```

```python
df.info()
```

Após mudar a coluna gastos_total para float verifiquei que tem 11 elementos nulos

```python
df[df['gastos_total'].isna()]
```

Os valores nulos da coluna gastos_total tem relação com tempo de contrato no valor 0. Possivelmente se trata de contratos novos, mas pra não ter duvida irei eliminar esse valores também.

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.head(5)
```

### Criando nova coluna de gastos diário

```python
gastos_diarios = df['gastos_total'] / (df['tempo_contrato'] * 30)
```

```python
gastos_diarios = gastos_diarios.round(2)
gastos_diarios
```

```python
df.insert(19, 'gastos_diarios', gastos_diarios)
```
